# 교육청 급식식단정보

```python
import pandas as pd
import re
diet = pd.read_csv('급식식단정보.csv')
diet.head()
```

```python
diet = diet[['요리명','칼로리정보','영양정보']]
diet['요리명'] = diet['요리명'].apply(lambda x: ', '.join(re.findall('[ㄱ-힣]+', x)))
diet['칼로리정보'] = diet['칼로리정보'].apply(lambda x: float(x.split()[0]))
diet['탄수화물(g)'] = diet['영양정보'].apply(lambda x: float(re.findall('탄수화물\(g\) : (.+?)<br/>',x)[0]))
diet['단백질(g)'] = diet['영양정보'].apply(lambda x: float(re.findall('단백질\(g\) : (.+?)<br/>',x)[0]))
diet['지방(g)'] = diet['영양정보'].apply(lambda x: float(re.findall('지방\(g\) : (.+?)<br/>',x)[0]))
diet['비타민A(R.E)'] = diet['영양정보'].apply(lambda x: float(re.findall('비타민A\(R\.E\) : (.+?)<br/>',x)[0]))
diet['리보플라빈(mg)'] = diet['영양정보'].apply(lambda x: float(re.findall('리보플라빈\(mg\) : (.+?)<br/>',x)[0]))
diet['칼슘(mg)'] = diet['영양정보'].apply(lambda x: float(re.findall('칼슘\(mg\) : (.+?)<br/>',x)[0]))
diet['철분(mg)'] = diet['영양정보'].apply(lambda x: float(re.findall('철분\(mg\) : (.+?)',x)[0]))
diet.drop(['영양정보'], axis=1, inplace=True)
diet.rename({'요리명':'식품명','칼로리정보':'에너지(kcal)'}, axis=1, inplace=True)
diet.head()
```

```python
diet.to_csv('diet.csv')
```
